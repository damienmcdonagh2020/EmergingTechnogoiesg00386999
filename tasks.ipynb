```python
def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence
```

```python
for i in range(1, 10001):
    sequence = collatz_sequence(i)
    print(f"Checking {i}: {sequence[-1]}")
    if sequence[-1] != 1:
        print(f"Collatz Conjecture not verified for {i}: {sequence}")
```

The Collatz conjecture, also known as the 3n+1 conjecture, is one of the most difficult to understand problems in maths. The conjecture posits that for any positive integer, if one repeatedly applies the function defined as dividing even numbers by 2 and tripling and adding 1 to odd numbers, the sequence will inevitably reach the repetitive pattern of 1, 4, 2, 1, 4, 2, and so on. While this conjecture has been extensively tested and holds true for an enormous range of positive integers, it has not been formally proven for all numbers, and thus, it stands as an unsolved puzzle in mathematics.
